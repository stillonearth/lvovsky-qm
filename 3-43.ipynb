```python
from sympy import *
from sympy.physics.quantum import *
from sympy.physics.quantum.qubit import *
```

```python
x, hbar, m, E, a, w0 = symbols('x hbar m E a w0')
psi = Function('psi')
```

```python
# V = - w0 * DiracDelta(x - a) - w0 * DiracDelta(x + a)
V = - w0 * DiracDelta(x)
```

```python
equation = Eq(-hbar**2/(2*m) * diff(psi(x), x, x) + V*psi(x), E*psi(x)); equation
```

```python
solution = dsolve(equation); solution
```
