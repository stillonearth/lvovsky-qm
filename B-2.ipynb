```python
import numpy as np
```

```python
W = 380e-9

def wave(z, t, A_H, A_V, phi_H, phi_V):
    c = 299792458.
    k = W / c
    i_comp = A_H * np.cos(k*z - W * t + phi_H)
    j_comp = A_V * np.cos(k*z - W * t + phi_V)
    return (i_comp, j_comp)
```

```python
wave(0, 0.001, 1, 0, 0, 0)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(A_H, A_V, phi_H, phi_V):
    # Data for plotting
    wt = np.arange(0.0, 2 * np.pi * 2, 0.01)
    t = wt / W
    Y = wave(0, t, A_H, A_V, phi_H, phi_V)
    
    fig, axs = plt.subplots(3)
    fig.suptitle('Wave vertical and horizontal')
    axs[0].plot(t, Y[0])
    axs[1].plot(t, Y[1])
    axs[2].plot(Y[0], Y[1])
```

```python
plot(1, 0, 0, 0)
```

```python
plot(5, -3, 0, 0)
```

```python
plot(5, -3, np.pi / 2, 0)
```

```python
plot(5, -3, np.pi / 4, - np.pi / 4)
```
